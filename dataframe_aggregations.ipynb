```python
import numpy as np
import pandas as pd
```

```python
new_df=pd.read_csv('./dataset.csv')
```

```python
new_df.head()
```

```python
new_df.info()
```

```python
new_df['Total Spent']=pd.to_numeric(new_df['Total Spent'],downcast='float',errors='coerce')
new_df['Price Per Unit']=pd.to_numeric(new_df['Price Per Unit'],downcast='float',errors='coerce')
new_df['Quantity']=pd.to_numeric(new_df['Quantity'],downcast='float',errors='coerce')

new_df.info()
```

```python
new_df['Quantity']*new_df['Price Per Unit']
```

# groupby()

```python
new_df.head()
```

```python
new_df.groupby('Item')[['Total Spent']].mean().sort_values('Total Spent',ascending=False)
```

# Grouping by multiple columns

```python
new_df.head()
```

```python
new_df.groupby(['Location','Item'],as_index=False)['Total Spent'].sum()
```

# Multi-index dataframes

```python
agg_new_df=new_df.groupby(['Location','Item'],as_index=True)['Total Spent'].sum()

agg_new_df
```

```python
agg_new_df.loc['In-store':'Takeaway',:]
```

```python
agg_new_df.loc[('In-store','Cake'):('Takeaway','Salad')]
```

# Modifying multi-index dataframes

```python
agg_new_df.reset_index()
```

```python
agg_new_df.swaplevel()
```

```python
agg_new_df.droplevel(0)
```

# agg method and named aggregations

```python
new_df.head()
```

```python
new_df.info()
```

```python
new_df.groupby(['Location','Item'],as_index=True)[['Total Spent']].agg(
    total_spent=('Total Spent','sum'),
    mean_spent=('Total Spent','mean')
)
```

# transform method

```python
new_df.head()
```

```python
agg_df=new_df.assign(
    avg_spent=new_df.groupby(['Item'])['Total Spent'].transform('mean'),
    difference=lambda x:x['Total Spent']-x['avg_spent']
)

agg_df
```

```python
agg_df.groupby('Item')['Total Spent'].sum()
```
