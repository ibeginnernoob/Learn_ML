```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

import numpy as np
import pandas as pd
```

```python
new_df=pd.read_csv('./dataset.csv')
```

```python
new_df.head()
```

```python
new_df.info()
```

```python
new_df['Total Spent']=pd.to_numeric(new_df['Total Spent'],downcast='float',errors='coerce')
new_df['Price Per Unit']=pd.to_numeric(new_df['Price Per Unit'],downcast='float',errors='coerce')
new_df['Quantity']=pd.to_numeric(new_df['Quantity'],downcast='float',errors='coerce')

new_df['Item']=new_df['Item'].astype('category')
new_df['Location']=new_df['Location'].astype('category')
new_df['Payment Method']=new_df['Payment Method'].astype('category')

new_df['Transaction Date']=pd.to_datetime(new_df['Transaction Date'],errors='coerce')

new_df.info()
```

```python
new_df['Quantity']*new_df['Price Per Unit']
```

# agg method and named aggregations

```python
new_df.head()
```

```python
new_df.info()
```

```python
new_df.groupby(['Location','Item'],as_index=True)[['Total Spent']].agg(
    total_spent=('Total Spent','sum'),
    mean_spent=('Total Spent','mean')
)
```

# transform method

```python
new_df.head()
```

```python
agg_df=new_df.assign(
    avg_spent=new_df.groupby(['Item'])['Total Spent'].transform('mean'),
    difference=lambda x:x['Total Spent']-x['avg_spent']
)

agg_df
```

```python
agg_df.groupby('Item')['Total Spent'].sum()
```

# Pivot tables

```python
new_df.info()
```

```python
new_df.head()
```

```python
new_df=new_df.assign(
    month=lambda x:x['Transaction Date'].dt.month
)

new_df['month']=pd.to_numeric(new_df['month'],downcast='float',errors='coerce')

new_df.info()
```

```python
new_df.pivot_table(
    index='month',
    columns='Item',
    values='Total Spent',
    aggfunc='sum'
)
```

```python
new_df.groupby(['month','Item'],as_index=False)[['Total Spent']].sum()
```
