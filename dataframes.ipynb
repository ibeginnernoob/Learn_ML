# Numpy select()

```python
conditions=[
    new_df['Item']=='Salad',
    new_df['Item']!='Salad'
]

choices=['Healthy','Not Healthy']

new_df['is_healthy']=np.select(conditions,choices,default='No Idea')

new_df
```

# map()

```python
new_df=pd.read_csv('./dataset.csv')
```

```python
new_df.head()
```

```python
new_df['Payment Method'].value_counts()
```

```python
online_dict={
    'Digital Wallet':'Online',
    'Credit Card':'Offline',
    'Cash':'Offline'
}

new_df['Payment Method']=new_df['Payment Method'].map(online_dict)

new_df
```

# assign()

```python
new_df=pd.read_csv('./dataset.csv')

new_df.head()
```

```python
new_df.assign(
    is_healthy_purchase=new_df['Item']=='Salad',
    tax_amount=pd.to_numeric(new_df['Total Spent'],errors='coerce')*0.18
).query('tax_amount>2')

# use lambda function to access newly created column from inside of the assign function
```

# Type Conversion

```python
new_df=pd.read_csv('./dataset.csv')

new_df.info()
```

```python
new_df['Quantity']=pd.to_numeric(new_df['Quantity'],errors="coerce")
new_df['Total Spent']=pd.to_numeric(new_df['Total Spent'],errors="coerce")

new_df.info()
```

```python
new_df=new_df.astype({'Quantity':'int','Total Spent':'int'},errors='ignore')

new_df
```
