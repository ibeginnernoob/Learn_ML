```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
plt.plot([0,1,4,9,16])
```

```python
data=[10*x for x in range(10)]
```

```python
labels=[f'{x}!' for x in range(10)]
```

```python
plt.plot(labels,data)
```

```python
df=pd.DataFrame(data,index=labels)

df=df.assign(series2=df[0]/2)

df.columns=['col1','col2']

df
```

```python
plt.plot(df)
```

# Object oriented approach for matplotlib

```python
fig,ax=plt.subplots()
ax.plot(data)

ax.set_title('Title')
fig.suptitle('Overall Title')
```

```python
fig,ax=plt.subplots()

ax.plot(df.index,df.loc[:,'col1'])
ax.plot(df.index,df.loc[:,'col2'],ls='--')

ax.set_title('Plotting Dataframe')
```
