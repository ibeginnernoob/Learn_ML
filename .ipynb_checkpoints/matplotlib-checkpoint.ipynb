# Chart Titles, adding legend and font size 

```python
new_df=pd.read_csv('./dataset.csv')

new_df.head()
```

```python
new_df['Total Spent']=pd.to_numeric(new_df['Total Spent'],errors='coerce')
new_df['Quantity']=pd.to_numeric(new_df['Quantity'],errors='coerce')
new_df['Price Per Unit']=pd.to_numeric(new_df['Price Per Unit'],errors='coerce')
new_df['Location']=new_df['Location'].astype('category')
new_df['Item']=new_df['Item'].astype('category')
new_df['Payment Method']=new_df['Payment Method'].astype('category')
new_df['Transaction Date']=pd.to_datetime(new_df['Transaction Date'],errors='coerce')

new_df.info()
```

```python
new_df.head()
```

```python
agg_df=new_df.groupby(['Transaction Date','Item'],as_index=False)['Total Spent'].sum()

agg_df=agg_df.assign(
    total_coffee_sales=lambda x: x['Total Spent'].where(x['Item']=='Coffee',0),
    total_salad_sales=lambda x: x['Total Spent'].where(x['Item']=='Salad',0)
).groupby('Transaction Date').agg({
    'total_coffee_sales':'sum',
    'total_salad_sales':'sum'
}).reset_index()

agg_df.head()
```

```python
fig,ax=plt.subplots()

ax.plot(agg_df['Transaction Date'],agg_df['total_coffee_sales'])
ax.plot(agg_df['Transaction Date'],agg_df['total_salad_sales'])
ax.set_xlabel('Day')
ax.set_ylabel('Daily Avg Spend')
ax.set_title('')

ax.legend(['Coffee Sales','Salad Sales'],ncol=2,loc='upper center')
```

# Line styles

```python
agg_df_with_month=agg_df.assign(
    month=agg_df['Transaction Date'].dt.month
)

final_agg_df=agg_df_with_month.groupby(['month']).agg(
    {
        'total_coffee_sales':'sum',
        'total_salad_sales':"sum"
    }
)

final_agg_df.head()
```

```python
fig,ax=plt.subplots()

ax.plot(
    final_agg_df.index,
    final_agg_df['total_coffee_sales'],
    linewidth=1,
    color='green',
    ls='--'
)
ax.plot(
    final_agg_df.index,
    final_agg_df['total_salad_sales'],
    linewidth=1,
    color='purple',
    ls=':'
)
ax.set_xlabel('Month of the year')
ax.set_ylabel('Avg Monthly Spend')
ax.set_title('')

ax.legend(['Coffee Sales','Salad Sales'],ncol=2,loc='center')
```
